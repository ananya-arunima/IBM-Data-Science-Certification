```python
# Problem 1
# Find the total number of crimes recorded in the CRIME table.

%sql Select count(*) from CHICAGO_CRIME_DATA

```

```python
# Problem 2
# List community areas with per capita income less than 11000

%sql select(COMMUNITY_AREA_NAME) from CENSUS_DATA \
where PER_CAPITA_INCOME < 11000
```

```python
# Problem 3 
# List all case numbers for crimes involving minors?(children are not considered minors 
# for the purposes of crime analysis)

%sql select CASE_NUMBER from CHICAGO_CRIME_DATA \
where DESCRIPTION LIKE '%MINOR%'
```

```python
%sql select CASE_NUMBER from CHICAGO_CRIME_DATA \
where DESCRIPTION LIKE '%CHILD%'
```

```python
# Problem 4
# List all kidnapping crimes involving a child?

%sql SELECT CASE_NUMBER, PRIMARY_TYPE, DATE, DESCRIPTION FROM CHICAGO_CRIME_DATA \
WHERE PRIMARY_TYPE LIKE'%KIDNAP%'
```

```python
# Problem 5¶
# What kinds of crimes were recorded at schools?

%sql SELECT DISTINCT(PRIMARY_TYPE), DESCRIPTION, LOCATION_DESCRIPTION FROM CHICAGO_CRIME_DATA \
WHERE LOCATION_DESCRIPTION LIKE'%SCHOOL%'
```

```python
# Problem 6
# List the average safety score for each type of school.

%sql SELECT "Elementary, Middle, or High School", AVG(SAFETY_SCORE) as AVERAGE_SAFETY_SCORE from \
CHICAGO_PUBLIC_SCHOOLS \
GROUP BY "Elementary, Middle, or High School";
```

```python
# Problem 7
# List 5 community areas with highest % of households below poverty line

%sql SELECT COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY from CENSUS_DATA \
ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC LIMIT 5
```

```python
# Problem 8
# Which community area is most crime prone?

%sql SELECT COMMUNITY_AREA_NUMBER, COUNT(COMMUNITY_AREA_NUMBER) as total_crime from CHICAGO_CRIME_DATA \
GROUP BY COMMUNITY_AREA_NUMBER \
ORDER BY COUNT(COMMUNITY_AREA_NUMBER) DESC LIMIT 1
```

```python
# Problem 9
# Use a sub-query to find the name of the community area with highest hardship index

%sql SELECT COMMUNITY_AREA_NAME FROM CENSUS_DATA \
WHERE HARDSHIP_INDEX = (SELECT MAX(HARDSHIP_INDEX) FROM CENSUS_DATA);
```

```python
# Problem 10
# Use a sub-query to determine the Community Area Name with most number of crimes?

%sql SELECT COMMUNITY_AREA_NUMBER from CHICAGO_CRIME_DATA \
WHERE COMMUNITY_AREA_NUMBER = (SELECT COMMUNITY_AREA_NUMBER FROM CHICAGO_CRIME_DATA \
                               GROUP BY COMMUNITY_AREA_NUMBER ORDER BY COUNT(COMMUNITY_AREA_NUMBER) \
                               DESC LIMIT 1) \
LIMIT 1
```
